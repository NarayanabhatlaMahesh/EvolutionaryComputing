````python
def GenEnvs(prompt):
    import os
    import subprocess
    def call_gemini(prompt):
        script_path = os.path.join(os.getcwd(), "PromptOptimizer.py")
    
        # print("Running:",         ["C:\\Users\\numam\\AppData\\Local\\Programs\\Python\\Python310\\python", script_path, prompt],)
    
        result = subprocess.run(
            ["C:\\Users\\numam\\AppData\\Local\\Programs\\Python\\Python310\\python", script_path, prompt],
            capture_output=True,
            text=True
        )
    
        return repr(result.stdout)
        # print("STDERR:", result.stderr)
    
        try:
            return json.loads(result.stdout)["response"]
        except json.JSONDecodeError:
            print("Failed to parse JSON from script output.")
            return None
    
    reply = call_gemini(prompt)
    # print(reply)
    return reply
````

````python
def call_for_Json_Check(prompt):
    import os
    import subprocess
    script_path = os.path.join(os.getcwd(), "CheckJsonLLM.py")

    result = subprocess.run(
        ["C:\\Users\\numam\\AppData\\Local\\Programs\\Python\\Python310\\python", script_path, 'json file:'+prompt],
        capture_output=True,
        text=True
    )

    return repr(result.stdout)

    try:
        return json.loads(result.stdout)["response"]
    except json.JSONDecodeError:
        print("Failed to parse JSON from script output.")
        return None
````

````python
import json
corrected_replies=[]
for i in Replies:
    try:
        val = call_for_Json_Check(i)
    except:
        print('\n\n\n\nsomeError\n\n\n\n')
        continue
    print(val)
    val = val[val.find('{'):]
    val=val[::-1]
    val = val[val.find('}'):][::-1]
    while(val.find('\\n')!=-1):
        val = val.replace('\\n','')
    print(val)
    while(val.find('\\')!=-1):
        val = val.replace('\\','')
    try:
        corrected_replies.append(json.loads(val.replace('\\\\n','').replace('```','').replace('"\\n\'','').replace('\'"json','')))
    except:
        print('error')
````

````python
for i,idx in enumerate(corrected_replies):
    print(idx['objects'],corrected_replies[i]['objects'].keys(),end='\n\n\n\n')
    if 'width' in corrected_replies[i].keys():
        corrected_replies[i]['grid_width']=corrected_replies[i]['width']
    if 'robot' in corrected_replies[i]['objects'].keys():
        print('\n\n\n\n\nrobot exists\n\n\n\n\n\n')
        corrected_replies[i]['objects'] = corrected_replies[i]['objects'].pop('robot',None)
````

````python
from evogym import EvoWorld, EvoSim, EvoViewer, sample_robot
import numpy as np
from numba import njit, prange

import os
base_dir = os.path.join("PromptOptimizationEnvGen", "initial_population")

# @njit(parallel=True)
def somefunc():
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            file_path = os.path.join(root, file)
            print(file_path)
            try:
                world = EvoWorld.from_json(file_path)
                world.pretty_print()
            except Exception as e:
                print('evowirld error',e)
                continue
            
            
            robot_structure, robot_connections = sample_robot((5, 5))
            try:
                world.add_from_array(
            	name='robot',
            	structure=robot_structure,
            	x=8,
            	y=18,
            	connections=robot_connections
            )
            except Exception as e:
                print('error',e)
                continue
            world.pretty_print()
            sim = EvoSim(world)
            sim.reset()
            viewer = EvoViewer(sim)
            viewer.track_objects('robot')
            for i in range(500):
            	sim.set_action(
            		'robot',
            		np.random.uniform(
            			low = 0.6,
            			high = 1.6,
            			size=(sim.get_dim_action_space('robot'),)
            		)
            	)
            	sim.step()
            	viewer.render('screen')
somefunc()
````
