```python
import pandas as pd
```

```python
file = pd.read_csv('LLMArtificialEcosystem.csv')
```

```python
file
```

```python
title = file['Abstract']
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="Falconsai/text_summarization")

Abstracts = [] 
for i in title:
    Abstracts.append(summarizer(i, max_length=1000, min_length=30, do_sample=False))
```
